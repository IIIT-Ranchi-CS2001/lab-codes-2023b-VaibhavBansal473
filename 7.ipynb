```python
import matplotlib.pyplot as plt

# Data for Madhya Pradesh
parties = ['BJP', 'INC', 'BSP', 'OTHERS']
madhyapradesh = [163, 66, 0, 1]
explode_mp = [0.05 if i == max(madhyapradesh) else 0 for i in madhyapradesh]

# Data for Rajasthan
rajasthan = [115, 69, 2, 13]
explode_rj = [0.05 if i == max(rajasthan) else 0 for i in rajasthan]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 7))

# Pie chart for Madhya Pradesh
ax1.pie(madhyapradesh, explode=explode_mp, labels=parties, autopct='%1.1f%%', shadow=False, startangle=180)
ax1.set_title('Madhya Pradesh')

# Pie chart for Rajasthan
ax2.pie(rajasthan, explode=explode_rj, labels=parties, autopct='%1.1f%%', shadow=False, startangle=180)
ax2.set_title('Rajasthan')

# Display the charts
plt.show()
# print(explode_mp)
```

```python
import matplotlib.pyplot as plt

# Data for Madhya Pradesh and Rajasthan
data_mp = {'BJP': 163, 'INC': 66, 'BSP': 0, 'Others': 1}
data_rj = {'INC': 69, 'BJP': 115, 'BSP': 2, 'Others': 13}

# Extract data for plotting
parties_mp = list(data_mp.keys())
seats_mp = list(data_mp.values())

parties_rj = list(data_rj.keys())
seats_rj = list(data_rj.values())

# Colors for better distinction
colors = ['gold', 'lightcoral', 'lightskyblue', 'lightgreen']

# Plot 1: Separate pie charts for Madhya Pradesh and Rajasthan
plt.figure(figsize=(12, 6))

# Madhya Pradesh pie chart
plt.subplot(121)
plt.pie(seats_mp, labels=parties_mp, autopct='%1.1f%%', startangle=140, colors=colors, explode=[0.1 if i == max(seats_mp) else 0 for i in seats_mp])
plt.title('Madhya Pradesh Assembly Elections 2023')

# Rajasthan pie chart
plt.subplot(122)
plt.pie(seats_rj, labels=parties_rj, autopct='%1.1f%%', startangle=140, colors=colors, explode=[0.1 if i == max(seats_rj) else 0 for i in seats_rj])
plt.title('Rajasthan Assembly Elections 2023')

plt.show()

# Plot 2: Subplots on the same figure object
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Madhya Pradesh subplot
axes[0].pie(seats_mp, labels=parties_mp, autopct='%1.1f%%', startangle=140, colors=colors, explode=[0.1 if i == max(seats_mp) else 0 for i in seats_mp])
axes[0].set_title('Madhya Pradesh Assembly Elections 2023')

# Rajasthan subplot
axes[1].pie(seats_rj, labels=parties_rj, autopct='%1.1f%%', startangle=140, colors=colors, explode=[0.1 if i == max(seats_rj) else 0 for i in seats_rj])
axes[1].set_title('Rajasthan Assembly Elections 2023')

plt.show()

# Plot 3: Bar chart with results for both states
fig, ax = plt.subplots(figsize=(12, 6))
x = range(len(parties_mp))
bar_width = 0.35

# Plot bars for Madhya Pradesh and Rajasthan
ax.bar(x, seats_mp, bar_width, label='Madhya Pradesh', color='lightblue')
ax.bar([i + bar_width for i in x], seats_rj, bar_width, label='Rajasthan', color='lightcoral')

# Adding details
ax.set_xlabel('Parties')
ax.set_ylabel('Seats Won')
ax.set_title('Assembly Elections 2023: Madhya Pradesh and Rajasthan')
ax.set_xticks([i + bar_width/2 for i in x])
ax.set_xticklabels(parties_mp)
ax.legend()

plt.show()
```
